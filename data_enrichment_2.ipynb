### remove special characters in the text column

```python
def remove_special_characters(text):
    # Replace all non-alphanumeric characters (except spaces) with an empty string
    return re.sub(r'[^A-Za-z0-9\s]+', '', text)

external_reviews_df['Text'] = external_reviews_df['Text'].apply(remove_special_characters)
```

### dropping unnecessary columns in the dataset

```python
# get the external_reviews_df without userId ,user location, user_locale,traveldate, published_date
external_reviews_df = external_reviews_df.drop(['User_ID', 'User_Location', 'User_Locale', 'Travel_Date', 'Published_Date', 'Helpful_Votes', 'Title'], axis=1)
external_reviews_df
```

### merging rows with same location

```python
merged_df = external_reviews_df.groupby(['Location_Name', 'Located_City', 'Location', 'Location_Type']).agg({
    'User_Contributions': 'sum',  # Sum the user contributions
    'Rating': 'mean',  # Average the rating
    'Text': lambda x: "', '".join(x)  # Concatenate reviews
}).reset_index()

merged_df['Text'] = "'" + merged_df['Text'].astype(str) + "'"

# Display the resulting DataFrame
print(merged_df)
```

```python
# get rows with null for rating on places_df
places_df[places_df['rating'].isnull()]['name']
```

```python
# Get the rows from 'places_df' where 'rating' is null
null_rating_names = places_df[places_df['rating'].isnull()]['name'].unique()

# Iterate over each unique name in places_df
for name in null_rating_names:
    # Search for each name in the Location_Name of merged_df
    matched_rows = merged_df[merged_df['Location_Name'].str.contains(name, case=False, na=False)]

    # Display the results
    if not matched_rows.empty:
        print(f"Matches for '{name}':")
        print(matched_rows)
        print("\n")
```
